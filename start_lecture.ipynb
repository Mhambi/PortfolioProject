```python
import matplotlib.pyplot as plt
number_of_datapoints=100
x=[0]*number_of_datapoints
y=[0]*number_of_datapoints

print(range(len(x)))

for i in range(len(x)):
  x[i]=i/10
  y[i]=i**2

plt.plot(x,y)
plt.show()



```

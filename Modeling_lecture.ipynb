```python
import matplotlib.pyplot as plt
import numpy as np
def gaussian (x, mu, sig):
  return(1.0/(np.sqrt(2.0*np.pi)*sig)*np.exp(-np.power((x-mu)/sig,2.0)/2))

max_t = 95
t=[24]*max_t
p_supply=[0]*max_t  # in [W] (J.s-1)
p_loads=[0]*max_t   # in [W] (J.s-1)
e_charge=[0]*max_t  # in [J]
dt=15*60            # in seconds

print(dt)
```

We have defined some variables, yaay ;)

```python
for i in range(len(t)-1):
  t[i]=i*0.25
  p_supply[i]=100*gaussian (i,50,10)*1000
  p_loads[i]=30000*gaussian(i,60,10)+5000*gaussian(i,80,3)
  e_charge[i+1]=e_charge[i]+(p_supply[i]-p_loads[i])*dt

plt.xlabel('Time[hr]')
plt.ylabel('Power[W]')
plt.plot(t,p_supply,t,p_loads)
plt.show()
plt.xlabel('Time[hr]')
plt.ylabel('Energy[J]')
plt.plot(t,e_charge)
plt.show()
```
